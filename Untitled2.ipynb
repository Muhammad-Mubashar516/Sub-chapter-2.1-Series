```python
import pandas as pd
import numpy as np

# ==========================================
# Chapter 2: Series (All Topics)
# ==========================================

print("=== Topic 1: Creating from Lists ===")
s1 = pd.Series([10, 20, 30])
print(s1)

print("\n=== Topic 2: Creating from Tuples ===")
s2 = pd.Series((40, 50, 60))
print(s2)

print("\n=== Topic 3: Creating from Dictionaries ===")
s3 = pd.Series({"A": 100, "B": 200, "C": 300})
print(s3)

print("\n=== Topic 4: Creating from NumPy Arrays ===")
arr = np.array([1, 2, 3, 4])
s4 = pd.Series(arr)
print(s4)

print("\n=== Topic 5: Creating from Scalars ===")
s5 = pd.Series(7, index=[0, 1, 2])
print(s5)

print("\n=== Topic 6: Custom Indexing ===")
s6 = pd.Series([11, 22, 33], index=["x", "y", "z"])
print(s6)

print("\n=== Topic 7: Default Indexing ===")
s7 = pd.Series([5, 10, 15])
print(s7)

print("\n=== Topic 8: Accessing by Label ===")
print(s6["y"])

print("\n=== Topic 9: Accessing by Position ===")
print(s6[1])

print("\n=== Topic 10: Slicing Series ===")
print(s7[0:2])

print("\n=== Topic 11: Arithmetic Addition ===")
print(s1 + s7)

print("\n=== Topic 12: Arithmetic Subtraction ===")
print(s1 - s7)

print("\n=== Topic 13: Arithmetic Multiplication ===")
print(s1 * 2)

print("\n=== Topic 14: Arithmetic Division ===")
print(s1 / 2)

print("\n=== Topic 15: Broadcasting Operations ===")
print(s1 + 5)

print("\n=== Topic 16: Handling NaN in Operations ===")
s8 = pd.Series([10, 20, np.nan])
print(s8.fillna(0))

print("\n=== Topic 17: Attribute: .values ===")
print(s1.values)

print("\n=== Topic 18: Attribute: .index ===")
print(s1.index)

print("\n=== Topic 19: Attribute: .dtype ===")
print(s1.dtype)

print("\n=== Topic 20: Attribute: .shape ===")
print(s1.shape)

print("\n=== Topic 21: Attribute: .ndim ===")
print(s1.ndim)

print("\n=== Topic 22: Method: .head() ===")
print(s3.head(2))

print("\n=== Topic 23: Method: .tail() ===")
print(s3.tail(2))

print("\n=== Topic 24: Method: .describe() ===")
print(s3.describe())

print("\n=== Topic 25: Method: .unique() ===")
s9 = pd.Series([1, 2, 2, 3, 3, 3])
print(s9.unique())

print("\n=== Topic 26: Method: .value_counts() ===")
print(s9.value_counts())

print("\n=== Topic 27: Vectorized String Operations ===")
s10 = pd.Series(["apple", "banana", "cat"])
print(s10.str.upper())

print("\n=== Topic 28: Alignment in Operations ===")
a = pd.Series([10, 20, 30], index=["A", "B", "C"])
b = pd.Series([5, 15, 25], index=["B", "C", "D"])
print(a + b)

print("\n=== Topic 29: Copying Series ===")
s11 = s1.copy()
print(s11)

print("\n=== Topic 30: Modifying In-Place ===")
s11[0] = 999
print(s11)
```
